# Steady State Error

```python
from IPython.core.display import HTML
import numpy as np

import matplotlib.pyplot as plt
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.size": 10,
})

from sympy import *
from sympy.plotting import plot

from mathprint import *
```

```python
t = symbols('t', real=True)
s = symbols('s', complex=True)
G, H, R, E, C, Ess = symbols('G H R E C E_ss', complex=True)
K = symbols("K", positive=True)
```

Define a system $G(s)$ with a feedback $H(s)$. The input is $r(t)$ and the output is $c(t)$.

![](./images/p23a.png)

Define the error:  

$e(t) = r(t) - c(t)$

and the steady state error:  

$e_{ss} = \lim\limits_{t \to \infty} e(t)$

With final value theorem:

$e_{ss} = \lim\limits_{s \to 0} sE(s)$


```python
eq1 = Eq(E, R - H*C)
eq2 = Eq(C, G*E)

sol = solve((eq1, eq2), (E, C))

mprint("E = ", latex(sol[E]))
mprint("C = ", latex(sol[C]))
```

In classical control theory, steady-state error is commonly evaluated using three standard test inputs: the __step input__, the __ramp input__, and the __parabolic input__.

## System Types


Define system type as the number of pure integrators in the open-loop transfer function $KG(s)H(s)$.

```python
GH_type0 = K
GH_type1 = K/s
GH_type2 = K/s**2

mprint("GH_{type0} = ", latex(GH_type0))
mprint("GH_{type1} = ", latex(GH_type1))
mprint("GH_{type2} = ", latex(GH_type2))
```

## Step Input

```python
eq3 = Subs(eq1, R, 1/s).evalf()
mprint(latex(eq3))

sol = solve((eq3, eq2), (E, C))
mprint("E = ", latex(sol[E]))

lim = limit(s*sol[E], s, 0)
mprint("E_{ss} = ", latex(lim))
```

## Ramp Input

```python
K = symbols("K", positive=True)

# Type 0
GH_type0 = K

E_type0 = 1/(s**2 * (1 + GH_type0))
ess_type0 = limit(s*E_type0, s, 0)

mprint("E_{type0}(s) = ", latex(E_type0))
mprint("e_{ss,type0} = ", latex(ess_type0))
```
